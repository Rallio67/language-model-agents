#Example Dialogue Feedback System 
This is some simple python code to easily allow a human to provide feedback for questions and answers in support of building an opensource AI assistant.

```python
import random

"""
Import QA tab separated text file where:
          column1 = question_id
          column2 = answer_a
          column3 = answer_b
"""

#open text file in read mode

with open("qa_file.txt") as text_file:
    questions = text_file.readlines()

text_file.close()
history=[]

def wrap_text(text, char_limit):
    # split the text into a list of words
    words = text.split()
    lines = []
    line = ""
    for word in words:
        # if the current line plus the next word would be too long,
        # append the current line to the list of lines and start a new line
        if len(line) + len(word) > char_limit:
            lines.append(line)
            line = ""
        # add the word to the current line
        line += " " + word
    # append the final line to the list of lines
    lines.append(line)
    # return the list of lines as a string,
    # with each line separated by a newline character
    return "\n   ".join(lines)


# Question Generator. Use a dictionary file from our website.

def get_question():
    entry=random.choice(questions)
    fields=entry.split("\t")
    dict_id=fields[0]
    question=fields[1]
    answer_a=fields[2]
    answer_b=fields[3]
    return(dict_id,question,answer_a,answer_b)

# Format text output.

def print_response(prompt):
  print("*"*50)
  print("-"*50)
  print("\n"+prompt[0]+"\n")
  print(wrap_text("(a)  "+prompt[1],80)[1:]+"\n")
  print(wrap_text("(b)  "+prompt[2],80)[1:]+"\n")
  print("(c) Both answers are bad.\n")
  print("(d) The question is flawed or can't be answered.\n")
  print("Note: type 'bye' to exit.")
# Define a function to handle player input

def handle_input(command):
    valid=False
    #print(history[-1:])
    if len(history) < 1 or command=="start":
        print("\nRosey: Hello, thanks for helping.\n\n\
Please do your best to answer the following questions!\n")
        dict_id,question,answer_a,answer_b=get_question()
        prompt=[question,answer_a,answer_b]
        print_response(prompt)
        return([dict_id,question,answer_a,answer_b,True,None])
    else:
        if history[-1] is None:
            history.pop(-1)
        
        past_result=history[-1:][0]
            
        if "a" == command or "b" == command or "c" == command or "d" == command and past_result[4]:
            response = "\nThank you!\nHere is another question.\n"
            print(response)
            dict_id,question,answer_a,answer_b=get_question()
            past_result[5]=command
            prompt=[question,answer_a,answer_b]
            print_response(prompt)
            return([dict_id,question,answer_a,answer_b,True,command])

        else:
          # If the player's input is not in the dictionary,
          # the NPC will provide a default response
            print("You must answer 'a' or 'b' or 'c' or 'd'\n\n")
        
```

```python
# Initialize the NPC's dialogue with the player
print("Rosey: Hello, human helper. Thank you for your willingness \
to grade my performance.\n\nPlease review my work and select which \
answer is best.\n\nType 'start' to begin.\n")

# Continuously prompt the player for input and handle
# their responses until they say "bye"
while True:
    player_input = input("Helper: ")
    if player_input.lower() == "bye":
        break
    history.append(handle_input(player_input))
# End the NPC's dialogue with the player
print("Rosey: Farewell, helper! I appreciate your assistance.")
history.pop(-1)
print("-"*50)
```
